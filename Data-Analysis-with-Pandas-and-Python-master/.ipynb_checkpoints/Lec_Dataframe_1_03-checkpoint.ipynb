## Fill in Null Values with the .fillna() Method

```python
import pandas as pd
```

```python
nba = pd.read_csv("nba.csv")
```

```python
nba.head(3)
```

```python
nba.fillna(0)
#works when you have a consistant data set
```

```python
nba["Salary"].fillna(0, inplace=True)
```

```python
nba.head()
```

```python
nba["College"].fillna("No College", inplace=True)
```

```python
nba.head()
```

## The .astype() Method

```python
nba.dropna(how = "all")
```

```python
nba["Salary"].fillna(0, inplace = True)
```

```python
nba["College"].fillna("None", inplace=True)
```

```python
nba.dtypes
nba.info()
```

```python
#convert float to int
nba["Salary"]= nba["Salary"].astype("int")
```

```python
nba.info()
```

```python
nba["Position"].nunique()
```

```python
nba["Postiion"] = nba["Position"].astype("category")
#category, a lot of vlaues in series have the same value
#use category instead of using string
```

## Sort a DataFrame with the .sort_values() Method. Part I

```python
nba = pd.read_csv("nba.csv")
```

```python
nba.head(3)
```

```python

nba.sort_values("Name", ascending = False)

nba.sort_values("Age", ascending = False)

nba.sort_values("Salary", ascending = False)
```

```python
nba.sort_values("Salary", na_position="first")
```

## Sort a DataFrame with the .sort_values() Method, Part II

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.sort_values(["Team", "Name"])
```

```python
nba.sort_values(["Team", "Name"], ascending = [True, False])
```
